# 1. Data Statistics

a. Data Kuantitatif (Usia)

```python
import pandas as pd

def analyze_age(file_path):
    # Baca data dari file Excel
    df = pd.read_excel(file_path)

    # Konversi kolom usia ke numerik
    df['Usia__dalam_tahun'] = pd.to_numeric(df['Usia__dalam_tahun'], errors='coerce')

    # Ambil data usia yang valid
    usia = df['Usia__dalam_tahun'].dropna()

    # Hitung statistik deskriptif
    stats = {
        'Rata-rata': usia.mean(),
        'Median': usia.median(),
        'Standar Deviasi': usia.std(),
        'Minimum': usia.min(),
        'Maksimum': usia.max(),
        'Kuartil 1': usia.quantile(0.25),
        'Kuartil 3': usia.quantile(0.75)
    }

    # Tampilkan hasil statistik
    print("Statistik Usia:")
    for k, v in stats.items():
        print(f"{k}: {v:.2f}")

# Jalankan analisis
if __name__ == '__main__':
    file_path = '18223037.xlsx'  # Ganti dengan path file yang sesuai
    analyze_age(file_path)
```

# 3. Confidence Intervals

a. One-sample confidence intervals for proportions

```python
import pandas as pd
from statsmodels.stats.proportion import proportion_confint

# Misalkan data sudah ada dalam variabel 'data'

# Kolom pertama
col1 = 'Seberapa_Setuju_Anda_dengan_pernyataan_berikut____Saya_merasakan_ketergantungan_pada_AI_untuk_menyelesaikan_tugas_tugas_tertentu_terutama_tugas_tugas_sederhana'
df['setuju_ai'] = df[col1].isin(['Setuju', 'Sangat Setuju']).astype(int)

# Hitung jumlah yang setuju
successes1 = df['setuju_ai'].sum()
n1 = df['setuju_ai'].count()  # Total responden
confidence_level = 0.95
ci_low1, ci_upp1 = proportion_confint(successes1, n1, alpha=1-confidence_level, method='normal')

print(f'Jumlah setuju untuk AI dependency: {successes1} dari {n1} responden')
print(f'One-sample confidence interval for AI dependency: ({ci_low1:.4f}, {ci_upp1:.4f})')

# Kolom kedua
col2 = 'Seberapa_Setuju_Anda_dengan_pernyataan_berikut____AI_sudah_memengaruhi_kehidupan_sehari_hari_saya'
df['setuju_ai_impact'] = df[col2].isin(['Setuju', 'Sangat Setuju']).astype(int)

# Hitung jumlah yang setuju
successes2 = df['setuju_ai_impact'].sum()
n2 = df['setuju_ai_impact'].count()  # Total responden
ci_low2, ci_upp2 = proportion_confint(successes2, n2, alpha=1-confidence_level, method='normal')

print(f'Jumlah setuju untuk AI impact: {successes2} dari {n2} responden')
print(f'One-sample confidence interval for AI impact: ({ci_low2:.4f}, {ci_upp2:.4f})')

# Insights berdasarkan hasil
insights = [
    "Jumlah responden yang merasakan ketergantungan pada AI untuk menyelesaikan tugas-tugas tertentu adalah {} dari {} responden.".format(successes1, n1),
    "Interval kepercayaan 95% untuk ketergantungan pada AI adalah ({:.4f}, {:.4f}).".format(ci_low1, ci_upp1),
    "Jumlah responden yang setuju bahwa AI sudah memengaruhi kehidupan sehari-hari mereka adalah {} dari {} responden.".format(successes2, n2),
    "Interval kepercayaan 95% untuk pengaruh AI adalah ({:.4f}, {:.4f}).".format(ci_low2, ci_upp2),
]

for insight in insights:
    print(insight)
```

b. Two-sample confidence intervals for proportion differences

```python
import pandas as pd
from statsmodels.stats.proportion import proportion_confint

# Misalkan data sudah ada dalam variabel 'data'

# Variabel pertama untuk analisis ketergantungan AI
col1 = 'Seberapa_Setuju_Anda_dengan_pernyataan_berikut____Saya_merasakan_ketergantungan_pada_AI_untuk_menyelesaikan_tugas_tugas_tertentu_terutama_tugas_tugas_sederhana'
df['setuju_ai'] = df[col1].isin(['Setuju', 'Sangat Setuju']).astype(int)

# Variabel kedua untuk analisis pengaruh AI
col2 = 'Seberapa_Setuju_Anda_dengan_pernyataan_berikut____AI_sudah_memengaruhi_kehidupan_sehari_hari_saya'
df['setuju_ai_impact'] = df[col2].isin(['Setuju', 'Sangat Setuju']).astype(int)

# Grouping berdasarkan jenis kelamin
group1 = df[df['Jenis_kelamin'] == 'Laki-laki']
group2 = df[df['Jenis_kelamin'] == 'Perempuan']

# Two-sample confidence interval for difference (ketergantungan AI)
group1_n1 = len(group1)
group1_successes1 = group1['setuju_ai'].sum()
group2_n1 = len(group2)
group2_successes1 = group2['setuju_ai'].sum()

if group1_n1 > 0 and group2_n1 > 0:
    p1 = group1_successes1 / group1_n1
    p2 = group2_successes1 / group2_n1
    diff1 = p1 - p2
    se_diff1 = ((p1 * (1 - p1) / group1_n1) + (p2 * (1 - p2) / group2_n1)) ** 0.5
    z = 1.96  # untuk 95% confidence level
    ci_low_diff1 = diff1 - z * se_diff1
    ci_upp_diff1 = diff1 + z * se_diff1

    print(f'Two-sample confidence interval for difference (AI dependency): ({ci_low_diff1:.4f}, {ci_upp_diff1:.4f})')
else:
    print("Salah satu kelompok tidak memiliki data yang cukup untuk analisis ketergantungan AI.")

# Two-sample confidence interval for difference (pengaruh AI)
group1_n2 = len(group1)
group1_successes2 = group1['setuju_ai_impact'].sum()
group2_n2 = len(group2)
group2_successes2 = group2['setuju_ai_impact'].sum()

if group1_n2 > 0 and group2_n2 > 0:
    p1 = group1_successes2 / group1_n2
    p2 = group2_successes2 / group2_n2
    diff2 = p1 - p2
    se_diff2 = ((p1 * (1 - p1) / group1_n2) + (p2 * (1 - p2) / group2_n2)) ** 0.5
    ci_low_diff2 = diff2 - z * se_diff2
    ci_upp_diff2 = diff2 + z * se_diff2

    print(f'Two-sample confidence interval for difference (AI impact): ({ci_low_diff2:.4f}, {ci_upp_diff2:.4f})')
else:
    print("Salah satu kelompok tidak memiliki data yang cukup untuk analisis pengaruh AI.")
```

# 4. Hypothesis Testing

a. One-sample z-test for proportions

```python
import pandas as pd
from statsmodels.stats.proportion import proportions_ztest

# Contoh: Inisialisasi data sudah ada pada variabel 'data' sebelumnya, misal setelah preprocessing

# Misal kolom persetujuan pada AI dependency
col = 'Seberapa_Setuju_Anda_dengan_pernyataan_berikut____Saya_merasakan_ketergantungan_pada_AI_untuk_menyelesaikan_tugas_tugas_tertentu_terutama_tugas_tugas_sederhana'
df['setuju_ai'] = df[col].isin(['Setuju', 'Sangat Setuju']).astype(int)

# One-sample z-test proporsi
successes = df['setuju_ai'].sum()
n = df['setuju_ai'].count()
p0 = 0.5  # proporsi hipotesis nol

z_stat, p_value = proportions_ztest(count=successes, nobs=n, value=p0)
print(f'One-sample z-test proporsi untuk kolom ketergantungan AI:')
print(f'Z-statistic: {z_stat:.4f}')
print(f'P-value: {p_value:.4f}')

# Two-sample z-test proporsi antara Laki-laki dan Perempuan
group1 = df[df['Jenis_kelamin'] == 'Laki-laki']
group2 = df[df['Jenis_kelamin'] == 'Perempuan']

successes_group1 = group1['setuju_ai'].sum()
n_group1 = group1['setuju_ai'].count()

successes_group2 = group2['setuju_ai'].sum()
n_group2 = group2['setuju_ai'].count()

count = [successes_group1, successes_group2]
nobs = [n_group1, n_group2]

z_stat2, p_value2 = proportions_ztest(count, nobs)
print('\nTwo-sample z-test proporsi antara Laki-laki dan Perempuan untuk ketergantungan AI:')
print(f'Z-statistic: {z_stat2:.4f}')
print(f'P-value: {p_value2:.4f}')


```

# 5. Chi-Square Tests

a. Goodness of Fit Test

```python
from scipy.stats import chisquare

# Hitung frekuensi yang diharapkan
observed = df['Pendidikan_terakhir'].value_counts()
expected = [len(df) / len(observed)] * len(observed)  # distribusi uniform

chi2_stat, p_value = chisquare(observed, expected)
alpha = 0.05  # atau 0.01 untuk 1%

print(f'Chi-Square Statistic (Goodness of Fit): {chi2_stat:.4f}, p-value: {p_value:.4f}')
if p_value < alpha:
    print("H0 ditolak: Distribusi variabel pilihan ≠ distribusi uniform.")
else:
    print("Gagal menolak H0: Distribusi variabel pilihan = distribusi uniform.")
```

b. Independence Test

```python
from scipy.stats import chi2_contingency

# Buat tabel kontingensi

contingency_table = pd.crosstab(df['Pekerjaan'], df['Domisili_saat_ini'])
chi2_stat, p_value, dof, expected = chi2_contingency(contingency_table)

print(f'Chi-Square Statistic (Independence Test): {chi2_stat:.4f}, p-value: {p_value:.4f}')
if p_value < alpha:
    print("H0 ditolak: Ada hubungan antara Pekerjaan dan Domisili.")
else:
    print("Gagal menolak H0: Tidak ada hubungan antara Pekerjaan dan Domisili.")
```

c. Homogeneity Test

```python
import pandas as pd
from scipy.stats import chi2_contingency

# Buat tabel kontingensi dengan variabel yang ada
contingency_table_homogeneity = pd.crosstab(
    df['Pendidikan_terakhir'],
    df['Domisili_saat_ini']  # Ganti dengan kolom yang ada
)

# Lakukan Chi-Square Test untuk Homogeneity
chi2_stat_homo, p_value_homo, dof_homo, expected_homo = chi2_contingency(contingency_table_homogeneity)

alpha = 0.05  # tingkat signifikansi 5%

print(f'Chi-Square Statistic (Homogeneity Test): {chi2_stat_homo:.4f}, p-value: {p_value_homo:.4f}')
if p_value_homo < alpha:
    print("H0 ditolak: Distribusi domisili tidak homogen antar pendidikan.")
else:
    print("Gagal menolak H0: Distribusi domisili homogen antar pendidikan.")
```

# 6. Kolmogorov-Smirnov Tests

```python
import pandas as pd
from scipy.stats import kstest

# Ambil kolom 'Usia'
usia = df['Usia__dalam_tahun'].dropna()  # Pastikan tidak ada nilai NaN

# Uji normalitas untuk kolom 'Usia'
statistic, p_value = kstest(usia, 'norm', args=(usia.mean(), usia.std()))

# Tentukan alpha
alpha = 0.05  # atau 0.01 untuk 1%

print(f'KS Statistic: {statistic:.4f}, p-value: {p_value:.4f}')

# Pemeriksaan H0
if p_value < alpha:
    print("H0 ditolak: Distribusi umur ≠ distribusi normal.")
else:
    print("Gagal menolak H0: Distribusi umur = distribusi normal.")
```

```python
print("--- INSIGHTS KOMPREHENSIF DARI ANALISIS DATA ---")

print("\n## 1. Statistik Usia (Analisis Kuantitatif)")
print("### Hasil Analisis:")
print(f"Rata-rata Usia: 34.80")
print(f"Median Usia: 21.00")
print(f"Standar Deviasi Usia: 113.99")
print(f"Usia Minimum: 11.00")
print(f"Usia Maksimum: 2008.00")
print(f"Kuartil 1 (25%): 19.00")
print(f"Kuartil 3 (75%): 36.00")
print("\n### Insights:")
print("- **Perbedaan Signifikan Rata-rata dan Median Menunjukkan Skewness Data:** Rata-rata usia responden adalah **34.80 tahun**, namun mediannya jauh lebih muda yaitu **21.00 tahun**. Perbedaan ini (rata-rata > median) secara kuat mengindikasikan bahwa distribusi usia **cenderung miring ke kanan (right-skewed)**.")
print("- **Adanya Outlier Ekstrem Menarik Rata-rata:** Kehadiran **usia maksimum 2008.00 tahun** adalah anomali data yang sangat jelas. Outlier ini yang secara drastis meningkatkan rata-rata usia dan standar deviasi (**113.99**), membuat rata-rata tidak merepresentasikan usia umum responden. Data ini perlu dibersihkan untuk analisis yang valid.")
print("- **Mayoritas Responden Berusia Muda Dewasa:** Terlepas dari outlier, nilai Kuartil 1 (**19.00 tahun**) dan Kuartil 3 (**36.00 tahun**) menunjukkan bahwa **50% responden berada dalam rentang usia muda dewasa hingga dewasa awal**.")

print("\n---")

print("\n## 2. Analisis Variabel Kualitatif (Modus, Frekuensi, Proporsi)")

print("\n### Pendidikan Terakhir")
print("Modus: SMA")
print("Proporsi SMA: 49.0%")
print("Proporsi S1: 32.3%")
print("Proporsi S2: 10.1%")
print("Proporsi SMP: 1.0%")
print("Proporsi SD: 0.3%")
print("\n### Insights:")
print("- **Dominasi Pendidikan Menengah Atas dan Sarjana:** Hampir separuh responden memiliki **pendidikan terakhir SMA (49.0%)**, diikuti oleh **lulusan S1 (32.3%)**. Ini mengindikasikan bahwa sampel survei sebagian besar terdiri dari individu dengan latar belakang pendidikan menengah hingga sarjana, yang mungkin relevan dengan topik AI.")
print("- **Representasi Pendidikan Rendah Sangat Minim:** Tingkat pendidikan SD dan SMP hanya diwakili oleh **0.3% dan 1.0%** responden, menunjukkan bahwa survei ini kurang menjangkau populasi dengan pendidikan dasar.")

print("\n### Pekerjaan")
print("Modus: Pelajar/Mahasiswa")
print("Proporsi Pelajar/Mahasiswa: 56.6%")
print("Proporsi Karyawan Swasta: 18.9%")
print("Proporsi Pegawai Negeri: 6.3%")
print("Proporsi Ibu Rumah Tangga: 6.0%")
print("\n### Insights:")
print("- **Mayoritas Responden adalah Pelajar/Mahasiswa:** **Lebih dari separuh responden (56.6%) adalah Pelajar/Mahasiswa**. Hal ini sangat konsisten dengan median usia yang muda (21 tahun) dan dominasi pendidikan SMA/S1, menyiratkan bahwa survei ini berhasil menangkap opini dari segmen demografi yang terpelajar dan mungkin sedang dalam fase pendidikan atau awal karir.")
print("- **Karyawan Swasta Menjadi Kelompok Pekerjaan Terbesar Kedua:** Dengan **18.9%**, karyawan swasta menjadi kelompok pekerjaan terbesar kedua, menunjukkan keragaman dalam sampel, meskipun tidak sebesar kelompok pelajar/mahasiswa.")

print("\n### Domisili Saat Ini")
print("Modus: Bandung")
print("Proporsi Bandung: 34.4%")
print("Proporsi Jakarta: 23.7%")
print("Jumlah Domisili Unik: 117")
print("\n### Insights:")
print("- **Fokus Sampel pada Kota Besar di Jawa Barat dan DKI Jakarta:** **Bandung (34.4%) dan Jakarta (23.7%)** adalah kota domisili mayoritas responden, secara kolektif menyumbang lebih dari separuh total responden. Ini mengindikasikan bahwa *insights* yang diperoleh mungkin lebih relevan untuk konteks perkotaan besar di Indonesia bagian barat.")
print("- **Cakupan Geografis yang Luas Namun Tidak Merata:** Meskipun ada **117 domisili unik**, proporsi responden dari daerah lain sangat kecil, menunjukkan survei memiliki cakupan geografis yang luas tetapi tidak merata.")

print("\n### Intensitas Aktivitas Online dalam 3 Bulan Terakhir")
print("Modus: Tinggi")
print("Proporsi Tinggi: 39.8%")
print("Proporsi Sangat Tinggi: 36.7%")
print("Total Proporsi Tinggi/Sangat Tinggi: 76.5%")
print("Proporsi Rendah/Sangat Rendah: 3.3%")
print("\n### Insights:")
print("- **Responden Sangat Aktif Digital:** Mayoritas responden menunjukkan **intensitas aktivitas online yang tinggi hingga sangat tinggi (total 76.5%)**. Ini adalah karakteristik penting dari sampel, menunjukkan bahwa mereka adalah pengguna internet yang aktif, sangat relevan untuk survei tentang teknologi seperti AI.")
print("- **Representasi Pengguna Online Pasif Sangat Minim:** Hanya sejumlah kecil responden (**3.3%**) yang melaporkan aktivitas online rendah atau sangat rendah, yang berarti survei ini tidak menangkap perspektif dari pengguna internet yang kurang aktif.")

print("\n### Tingkat Pengetahuan tentang Kecerdasan Buatan (AI)")
print("Modus: Cukup mengetahui")
print("Proporsi Cukup mengetahui: 65.1%")
print("Proporsi Sangat mengetahui dan paham: 23.4%")
print("Proporsi Pernah mendengar, tidak mengetahui detilnya: 11.2%")
print("Proporsi Belum pernah mendengar sama sekali: 0.3%")
print("\n### Insights:")
print("- **Kesadaran dan Pemahaman AI yang Tinggi:** Sebagian besar responden memiliki setidaknya pemahaman dasar tentang AI; **65.1% 'Cukup mengetahui'** dan **23.4% 'Sangat mengetahui dan paham'**. Ini berarti **sekitar 88.5% responden memiliki setidaknya pengetahuan yang cukup tentang AI**.")
print("- **AI Bukan Lagi Istilah Asing:** Sangat sedikit responden (**0.3%**) yang 'belum pernah mendengar sama sekali', mengindikasikan bahwa istilah Kecerdasan Buatan sudah sangat dikenal di kalangan populasi yang disurvei.")

print("\n---")

print("\n## 3. Visualisasi Data (Interpretasi Berdasarkan Jenis Plot yang Dibuat)")
print("### Insights:")
print("\n### Histogram Usia")
print("- **Distribusi Usia Sangat Miring ke Kanan:** Histogram menunjukkan puncak frekuensi yang sangat tinggi pada rentang usia **15-25 tahun**, khususnya sekitar **20-22 tahun**, dan kemudian menurun drastis. Ini secara visual mengkonfirmasi bahwa **mayoritas responden berusia sangat muda**.")
print("- **Adanya Outlier Usia Lanjut/Anomali:** Meskipun puncaknya di usia muda, ada *bar* kecil yang muncul di usia **45-50 tahun ke atas**, bahkan hingga sekitar **50-55 tahun**. Ini mengindikasikan keberadaan responden yang lebih tua, atau potensi *outlier* yang menyebabkan distribusi menjadi miring, seperti yang disebutkan dalam statistik.")
print("\n### Distribusi Usia Responden (KDE Plot) (Screenshot 2025-06-21 223225.png)")
print("### Insights:")
print("- **Konfirmasi Distribusi Usia yang Miring dan Puncak di Usia Muda:** KDE Plot secara jelas menunjukkan **puncak kepadatan yang tajam pada rentang usia 20-25 tahun**, yang kemudian menurun secara bertahap dan memanjang ke kanan. Ini memperkuat temuan histogram bahwa mayoritas responden berada di usia muda dan distribusi usia tidak normal (miring).")
print("- **Indikasi Adanya Kelompok Usia Lebih Tua yang Lebih Kecil:** Meskipun dominan pada usia muda, adanya 'ekor' yang memanjang dan sedikit *bump* di sekitar usia 40-50 tahun menunjukkan bahwa ada juga sekelompok kecil responden di usia yang lebih matang, meskipun frekuensinya jauh lebih rendah.")

print("\n### Jam Penggunaan AI Berdasarkan Pendidikan")
print("### Insights:")
print("- **Penggunaan AI Terkonsentrasi pada 'Kurang dari 1 jam' dan '1-2 jam' di Berbagai Tingkat Pendidikan:** Untuk mayoritas kategori pendidikan yang memiliki data (SMA, S1, S2, D3, D2, S3, SMK, SMP, D1), median jam penggunaan AI cenderung berada pada kategori **'Kurang dari 1 jam' atau '1-2 jam'**. Ini menunjukkan bahwa sebagian besar responden, terlepas dari tingkat pendidikannya, umumnya menghabiskan waktu singkat dalam sehari menggunakan AI.")
print("- **Variasi Luas dalam Penggunaan AI di Tingkat SMA, S1, dan S2:** Kategori SMA, S1, dan S2 menunjukkan rentang penggunaan AI yang paling luas, dari 'Kurang dari 1 jam' hingga 'Lebih dari 4 jam', dan bahkan beberapa *outlier* di '2-4 jam' atau 'Lebih dari 4 jam'. Ini mengindikasikan keragaman dalam kebiasaan penggunaan AI di antara kelompok pendidikan utama ini.")
print("- **Data Penggunaan AI yang Terbatas di Beberapa Tingkat Pendidikan:** Beberapa kategori pendidikan (SD, Nurse, D4, Apoteker) memiliki data yang sangat sedikit atau bahkan tidak ada kotak, sehingga sulit untuk menarik kesimpulan yang kuat mengenai pola penggunaan AI mereka.")

print("\n### Distribusi Usia Berdasarkan Penghasilan per Bulan (Violin Plot)")
print("### Insights:")
print("- **Distribusi Usia Paling Terkonsentrasi pada Kelompok Penghasilan '< Rp 2 juta':** Violin plot untuk '< Rp 2 juta' menunjukkan kepadatan usia yang sangat tinggi dan sempit di sekitar **15-25 tahun**, dengan median sekitar **20 tahun**. Hal ini sangat konsisten dengan data demografi bahwa mayoritas responden adalah pelajar/mahasiswa yang cenderung memiliki penghasilan rendah.")
print("- **Rentang Usia Lebih Luas pada Kelompok Penghasilan Lebih Tinggi:** Kelompok penghasilan di atas 'Rp 2 juta' (misalnya, 'Rp 5 juta - Rp 10 juta', 'Rp 10 juta - Rp 20 juta', '> Rp 20 juta', dan 'Rp 2 juta - Rp 5 juta') menunjukkan **distribusi usia yang jauh lebih lebar**, seringkali meliputi usia dari sekitar 15 tahun hingga di atas 60 tahun. Ini menunjukkan bahwa kelompok penghasilan yang lebih tinggi dihuni oleh individu dari berbagai rentang usia.")
print("- **Median Usia Cenderung Meningkat Seiring Penghasilan (Meski Tidak Monoton):** Meskipun tidak ada pola linear yang ketat, ada indikasi bahwa median usia cenderung sedikit bergeser ke atas untuk kelompok penghasilan yang lebih tinggi dibandingkan dengan kelompok '< Rp 2 juta'.")

print("\n### Usia Berdasarkan Aktivitas Akses Media Sosial")
print("### Insights:")
print("- **Sebagian Besar Responden Muda Berinteraksi dengan Aktivitas Media Sosial '1':** Untuk 'Aktivitas Akses Media Sosial' bernilai '1', terlihat **konsentrasi padat titik-titik (responden) pada rentang usia 18-25 tahun**. Ini menunjukkan bahwa aktivitas media sosial yang diwakili oleh '1' didominasi oleh kelompok usia muda.")
print("- **Responden dengan Aktivitas Media Sosial '0' Memiliki Rentang Usia Lebih Luas:** Untuk 'Aktivitas Akses Media Sosial' bernilai '0', meskipun jumlah titiknya lebih sedikit, sebarannya mencakup rentang usia yang lebih luas, dari sekitar 15 tahun hingga lebih dari 50 tahun. Ini mungkin menunjukkan bahwa aktivitas '0' ini dilakukan oleh berbagai kelompok usia, atau kurang spesifik untuk usia muda.")
print("- **Kebutuhan Akan Label yang Jelas untuk 'Aktivitas Akses Media Sosial':** Tanpa label yang jelas untuk '0' dan '1' pada sumbu X, sulit untuk menarik kesimpulan yang lebih spesifik tentang jenis aktivitas media sosial mana yang mendominasi kelompok usia tertentu. Asumsi awal adalah '1' mungkin merepresentasikan aktivitas media sosial yang lebih umum atau sering.")

print("\n### Rata-rata Usia Berdasarkan Sumber Mengetahui AI")
print("### Insights:")
print("- **Sumber 'Google/Artikel/Kelas' Diketahui oleh Kelompok Usia Paling Tua:** Sumber 'Google, Artikel, Kelas' memiliki rata-rata usia responden tertinggi, mencapai sekitar **50 tahun**. Ini menunjukkan bahwa kelompok usia yang lebih matang cenderung memperoleh informasi tentang AI dari sumber-sumber yang lebih formal atau mendalam.")
print("- **Sumber 'Media Sosial/Artikel' Diketahui oleh Kelompok Usia Termuda:** Sumber 'Media sosial, Artikel' memiliki rata-rata usia yang cenderung lebih rendah (sekitar **23-29 tahun**), yang konsisten dengan penggunaan media sosial yang lebih tinggi di kalangan anak muda.")
print("- **Berbagai Sumber Dikenal oleh Rentang Usia yang Berbeda:** Ada variasi rata-rata usia yang signifikan antar sumber, mulai dari yang muda (misalnya, 'Kehidupan sehari-hari' sekitar 19 tahun) hingga yang lebih tua. Hal ini mengindikasikan bahwa cara orang mengenal AI sangat dipengaruhi oleh kelompok usia mereka.")

print("\n### Usia vs Rata-rata Jam Penggunaan Internet per Hari")
print("### Insights:")
print("- **Tidak Ada Korelasi Linier yang Jelas antara Usia dan Jam Penggunaan Internet:** Scatter plot ini menunjukkan **sebaran titik yang acak** tanpa pola naik atau turun yang jelas. Ini mengindikasikan **tidak adanya korelasi linier kuat** antara usia responden dengan rata-rata jam penggunaan internet per hari mereka.")
print("- **Penggunaan Internet Berjam-jam Terjadi di Berbagai Usia:** Meskipun tidak ada korelasi, terlihat bahwa baik responden muda (sekitar 15-25 tahun) maupun yang lebih tua (hingga 50 tahun) sama-sama melaporkan jam penggunaan internet yang bervariasi, dari '1 jam' hingga '> 8 jam'. Ini menunjukkan bahwa durasi penggunaan internet tidak secara langsung bergantung pada usia dalam sampel ini.")
print("- **Konsentrasi Pengguna Internet di Usia Muda:** Meskipun sebarannya luas, ada sedikit konsentrasi titik di bagian bawah grafik (usia muda) yang juga melaporkan jam penggunaan internet yang tinggi, terutama di sekitar usia 20-25 tahun.")

print("\n---")

print("\n---")

print("\n## 4. Confidence Interval untuk Proporsi")

print("\n### Ketergantungan pada AI (One-Sample Confidence Interval)")
print(f"Jumlah setuju untuk AI dependency: 185 dari 1000 responden.")
print(f"One-sample confidence interval for AI dependency: (0.1609, 0.2091).")
print("\n### Insights:")
print("- **Ketergantungan AI Terasa oleh Seperlima Responden:** Sebanyak **185 dari 1000 responden (18.5%) merasakan ketergantungan pada AI** untuk menyelesaikan tugas-tugas tertentu, terutama yang sederhana.")
print("- **Interval Kepercayaan Kuat Menegaskan Proporsi Ketergantungan:** Dengan interval kepercayaan 95% sebesar **(0.1609, 0.2091)**, kita 95% yakin bahwa proporsi populasi sebenarnya yang merasakan ketergantungan AI berada antara **16.09% dan 20.91%**. Ini menunjukkan bahwa AI sudah menjadi alat bantu penting bagi sebagian signifikan responden dalam menyelesaikan tugas.")

print("\n### Pengaruh AI pada Kehidupan Sehari-hari (One-Sample Confidence Interval)")
print(f"Jumlah setuju untuk AI impact: 332 dari 1000 responden.")
print(f"One-sample confidence interval for AI impact: (0.3028, 0.3612).")
print("\n### Insights:")
print("- **Dampak AI Dirasakan oleh Sepertiga Responden:** Sebanyak **332 dari 1000 responden (33.2%) setuju bahwa AI sudah memengaruhi kehidupan sehari-hari mereka**.")
print("- **Interval Kepercayaan Menegaskan Persepsi Dampak AI yang Luas:** Dengan interval kepercayaan 95% sebesar **(0.3028, 0.3612)**, kita 95% yakin bahwa proporsi populasi yang merasa AI memengaruhi kehidupan mereka berada di kisaran **30.28% hingga 36.12%**. Ini adalah persentase yang signifikan, menegaskan dampak AI yang semakin meluas dan tidak hanya terbatas pada tugas sederhana.")

print("\n### Perbandingan Ketergantungan AI (Two-Sample Confidence Interval - Laki-laki vs Perempuan)")
print(f"Two-sample confidence interval for difference (AI dependency): (-0.0312, 0.0656).")
print("\n### Insights:")
print("- **Tidak Ada Perbedaan Gender Signifikan dalam Ketergantungan AI:** Interval kepercayaan 95% untuk perbedaan proporsi ketergantungan AI antara laki-laki dan perempuan adalah **(-0.0312, 0.0656)**. Karena interval ini **mencakup nilai nol**, kita dapat menyimpulkan bahwa **tidak ada perbedaan yang signifikan secara statistik** dalam tingkat ketergantungan pada AI antara laki-laki dan perempuan dalam sampel ini pada tingkat kepercayaan 95%.")

print("\n### Perbandingan Pengaruh AI (Two-Sample Confidence Interval - Laki-laki vs Perempuan)")
print(f"Two-sample confidence interval for difference (AI impact): (0.0661, 0.1832).")
print("\n### Insights:")
print("- **Laki-laki Cenderung Lebih Merasakan Pengaruh AI dalam Kehidupan Sehari-hari:** Interval kepercayaan 95% untuk perbedaan proporsi pengaruh AI antara laki-laki dan perempuan adalah **(0.0661, 0.1832)**. Karena interval ini **tidak mencakup nilai nol**, ada **perbedaan yang signifikan secara statistik** dalam persepsi pengaruh AI dalam kehidupan sehari-hari antara laki-laki dan perempuan. Nilai positif menunjukkan bahwa **proporsi laki-laki yang setuju AI memengaruhi kehidupan sehari-hari mereka lebih tinggi dibandingkan perempuan**.")

print("\n---")

print("\n## 5. Uji Hipotesis Proporsi (Z-test)")

print("\n### One-sample Z-test Proporsi untuk Ketergantungan AI")
print(f"Z-statistic: -25.6535")
print(f"P-value: 0.0000")
print("\n### Insights:")
print("- **Proporsi Ketergantungan AI Jauh Berbeda dari 50%:** Dengan Z-statistic **-25.6535** dan p-value **0.0000** (jauh lebih kecil dari alpha 0.05), kita **menolak hipotesis nol (H0)** bahwa proporsi responden yang merasakan ketergantungan AI adalah 0.5 (50%). Ini mengkonfirmasi bahwa proporsi ketergantungan AI **secara signifikan berbeda dari 50%**, sejalan dengan angka 18.5% yang diamati.")

print("\n### Two-sample Z-test Proporsi untuk Ketergantungan AI (Laki-laki vs Perempuan)")
print(f"Z-statistic: 0.6980")
print(f"P-value: 0.4852")
print("\n### Insights:")
print("- **Tidak Ada Bukti Perbedaan Signifikan Gender dalam Ketergantungan AI:** Dengan Z-statistic **0.6980** dan p-value **0.4852** (lebih besar dari alpha 0.05), kita **gagal menolak hipotesis nol (H0)**. Ini semakin menguatkan kesimpulan dari *confidence interval* bahwa **tidak ada perbedaan yang signifikan secara statistik** dalam proporsi ketergantungan AI antara laki-laki dan perempuan.")

print("\n---")

print("\n## 6. Uji Hipotesis Chi-Square")

print("\n### Chi-Square Test (Goodness of Fit) untuk Pendidikan Terakhir")
print(f"Chi-Square Statistic: 3654.8320")
print(f"P-value: 0.0000")
print(f"Kesimpulan: H0 ditolak: Distribusi variabel pilihan ≠ distribusi uniform.")
print("\n### Insights:")
print("- **Distribusi Pendidikan Tidak Seragam:** Dengan Chi-Square Statistic **3654.8320** dan p-value **0.0000** (jauh di bawah 0.05), kita **menolak hipotesis nol (H0)**. Ini secara statistik membuktikan bahwa **distribusi pendidikan terakhir responden tidak seragam**, yang sesuai dengan observasi bahwa kategori SMA dan S1 sangat dominan.")

print("\n### Chi-Square Test (Independence Test) untuk Pekerjaan dan Domisili")
print(f"Chi-Square Statistic: 8950.9836")
print(f"P-value: 0.0000")
print(f"Kesimpulan: H0 ditolak: Ada hubungan antara Pekerjaan dan Domisili.")
print("\n### Insights:")
print("- **Pekerjaan dan Domisili Saling Berhubungan:** Dengan Chi-Square Statistic **8950.9836** dan p-value **0.0000** (jauh di bawah 0.05), kita **menolak hipotesis nol (H0)**. Ini membuktikan bahwa **ada hubungan yang signifikan secara statistik antara pekerjaan dan domisili responden**. Artinya, jenis pekerjaan seseorang tidak independen dari tempat tinggalnya, dan ada pola tertentu antara jenis pekerjaan dan kota domisili.")

print("\n### Chi-Square Test (Homogeneity Test) untuk Pendidikan Terakhir dan Domisili")
print(f"Chi-Square Statistic: 1591.6127")
print(f"P-value: 0.0001")
print(f"Kesimpulan: H0 ditolak: Distribusi domisili tidak homogen antar pendidikan.")
print("\n### Insights:")
print("- **Distribusi Domisili Berbeda Antar Kelompok Pendidikan:** Dengan Chi-Square Statistic **1591.6127** dan p-value **0.0001** (jauh di bawah 0.05), kita **menolak hipotesis nol (H0)**. Ini menunjukkan bahwa **distribusi domisili responden tidak homogen antar kelompok pendidikan yang berbeda**. Artinya, proporsi individu dari domisili tertentu bervariasi secara signifikan tergantung pada tingkat pendidikan mereka, menyiratkan pengaruh pendidikan terhadap pilihan domisili atau sebaliknya.")

print("\n### Kolmogorov-Smirnov Test (Normalitas Usia)")
print(f"KS Statistic: 0.2600")
print(f"P-value: 0.0004")
print(f"Kesimpulan: H0 ditolak: Distribusi umur ≠ distribusi normal.")
print("\n### Insights:")
print("- **Distribusi Usia Tidak Normal secara Statistik:** Dengan KS Statistic **0.2600** dan p-value **0.0004** (jauh lebih kecil dari alpha 0.05), kita **menolak hipotesis nol (H0)**. Ini secara statistik mengkonfirmasi bahwa **distribusi usia responden tidak mengikuti distribusi normal**, konsisten dengan observasi awal adanya outlier dan skewness.")

print("\n---")

print("\n## Rekomendasi Utama & Langkah Selanjutnya")
print("Berdasarkan seluruh analisis yang telah dilakukan, berikut adalah rekomendasi utama dan langkah-langkah selanjutnya yang dapat diambil:")
print("1.  **Prioritas Utama: Pembersihan dan Validasi Data Usia.** Keberadaan nilai **'2008 tahun'** sebagai usia responden adalah anomali serius yang sangat memengaruhi semua analisis statistik deskriptif dan inferensial terkait usia. Lakukan pembersihan data dengan mengidentifikasi dan menghapus/mengoreksi entri usia yang tidak valid ini. Setelah pembersihan, ulangi analisis statistik deskriptif dan uji normalitas untuk kolom usia untuk mendapatkan gambaran yang lebih akurat.")
print("2.  **Fokus pada Demografi Mayoritas:** Mengingat **mayoritas responden adalah Pelajar/Mahasiswa (56.6%)** dengan **pendidikan terakhir SMA (49.0%) dan S1 (32.3%)**, *insight* yang didapat kemungkinan besar lebih representatif untuk kelompok demografi ini. Pertimbangkan untuk melakukan analisis segmentasi yang lebih mendalam untuk kelompok ini, atau jika tujuan survei adalah populasi yang lebih luas, perhatikan potensi bias sampel.")
print("3.  **Memahami Ketergantungan dan Pengaruh AI yang Berbeda:** Meskipun **tidak ada perbedaan signifikan gender dalam merasakan ketergantungan AI** untuk tugas sederhana, **ada perbedaan signifikan dalam persepsi bahwa AI memengaruhi kehidupan sehari-hari (laki-laki > perempuan)**. Penelitian kualitatif lebih lanjut mungkin diperlukan untuk memahami akar perbedaan persepsi ini.")
print("4.  **Implikasi Tingkat Pengetahuan AI yang Tinggi:** Dengan **88.5% responden setidaknya cukup mengetahui AI**, ini menunjukkan bahwa edukasi dasar tentang AI mungkin tidak lagi menjadi prioritas utama. Fokus dapat dialihkan ke pemahaman yang lebih dalam tentang aplikasi spesifik, etika, atau potensi AI.")
print("5.  **Korelasi Demografi dengan Perilaku:** Temuan adanya **hubungan signifikan antara Pekerjaan dan Domisili**, serta **Pendidikan Terakhir dan Domisili**, mengisyaratkan bahwa faktor geografis dan status sosial ekonomi saling terkait. Analisis multivariat lebih lanjut (misalnya, regresi logistik) dapat dilakukan untuk memodelkan bagaimana kombinasi demografi ini memengaruhi variabel minat lainnya.")
print("6.  **Pemanfaatan Tingginya Aktivitas Online:** Mayoritas responden (**76.5%**) memiliki intensitas aktivitas online tinggi hingga sangat tinggi. Ini adalah peluang besar untuk diseminasi informasi atau interaksi terkait AI melalui platform digital.")
```
